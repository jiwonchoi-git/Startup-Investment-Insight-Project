```python
from selenium import webdriver
from urllib.parse import quote_plus
from bs4 import BeautifulSoup as bs
import time
```

```python
baseurl = 'https://www.google.com/search?q='
plusurl = input('검색어: ')
url = baseurl + quote_plus(plusurl)  # 검색어를 URL 인코딩

driver = webdriver.Chrome()  # Chrome Driver의 경로를 지정해야 할 수도 있습니다.
driver.get(url)

html = driver.page_source
soup = bs(html, 'html.parser')

# 최신 구글 검색 결과 페이지에 맞게 선택자 업데이트
results = soup.find_all('div', class_='tF2Cxc')

# 관심 있는 키워드 목록
keywords = ['invest', 'venture', 'startup', 'seed', 'series']

for i in results:
    title = i.find('h3').text if i.find('h3') else 'No title'
    link = i.find('a')['href'] if i.find('a') else 'No link'
    # 제목 또는 링크에 키워드가 포함되어 있는지 확인
    if any(keyword in title.lower() for keyword in keywords) or any(keyword in link.lower() for keyword in keywords):
        print(title)
        print(link)
        print()

driver.close()
```

```python
from selenium import webdriver
from urllib.parse import quote_plus
from bs4 import BeautifulSoup as bs
import pandas as pd  # pandas 라이브러리 추가
import time  # 각 검색 사이에 딜레이를 주기 위해 추가
```

```python
baseurl = 'https://www.google.com/search?q='

# 엑셀 파일에서 기업 목록을 읽어오기
companies_df = pd.read_excel('C:/Users/user/companies.xlsx')  # 엑셀 파일의 경로 지정
companies = companies_df.iloc[:, 0].tolist()  # 첫 번째 열의 데이터를 리스트로 변환

driver = webdriver.Chrome()  # Chrome Driver의 경로를 지정해야 할 수도 있습니다.

# 검색 결과가 없는 기업들의 이름을 저장할 리스트
companies_with_no_results = []

for company in companies:
    search_query = company + ' investment'  # 검색어 설정
    url = baseurl + quote_plus(search_query)
    driver.get(url)
    time.sleep(5)  # 페이지 로딩 대기

    # 스크롤을 6번 내림
    for _ in range(6):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)  # 페이지가 추가 내용을 로드할 시간을 제공

    html = driver.page_source
    soup = bs(html, 'html.parser')

    results = soup.find_all('div', class_='tF2Cxc')

    # 관심 있는 키워드 목록
    keywords = ['venture', 'startup', 'seed', 'series']

    found = False  # 검색 결과 존재 여부를 추적하는 플래그

    for i in results:
        title = i.find('h3').text if i.find('h3') else 'No title'
        link = i.find('a')['href'] if i.find('a') else 'No link'
        if any(keyword in title.lower() for keyword in keywords) or any(keyword in link.lower() for keyword in keywords):
            print(f"Company: {company}")
            print(title)
            print(link)
            print()
            found = True

    if not found:
        companies_with_no_results.append(company)

    print('-----------------------------')

    time.sleep(2)  # 검색 제한 방지 대기 시간

driver.close()

# 검색 결과가 없는 기업들의 이름 출력
if companies_with_no_results:
    print("검색 결과가 없는 기업들:")
    for company in companies_with_no_results:
        print(company)
```

```python
baseurl = 'https://www.google.com/search?q='

# 엑셀 파일에서 기업 목록을 읽어오기
companies_df = pd.read_excel('C:/Users/user/companies.xlsx')  # 엑셀 파일의 경로 지정
companies = companies_df.iloc[:, 0].tolist()  # 첫 번째 열의 데이터를 리스트로 변환

driver = webdriver.Chrome()  # Chrome Driver의 경로를 지정해야 할 수도 있습니다.

# 검색 결과가 없는 기업들의 이름을 저장할 리스트
companies_with_no_results = []
companies_with_cvc = []

for company in companies:
    search_query = company + ' venture'
    url = baseurl + quote_plus(search_query)
    driver.get(url)
    time.sleep(5)  # Allow time for the page to load

    # Scroll down the page 6 times
    for _ in range(6):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)  # Allow time for additional content to load

    html = driver.page_source
    soup = bs(html, 'html.parser')

    results = soup.find_all('div', class_='tF2Cxc')
    found = False  # Flag to track if results are found

    # Process each result
    for i in results:
        title = i.find('h3').text if i.find('h3') else 'No title'
        link = i.find('a')['href'] if i.find('a') else 'No link'

        # Check if the title or link contains any of the keywords
        if any(keyword in title.lower() for keyword in keywords) or any(keyword in link.lower() for keyword in keywords):
            companies_with_cvc.append(company)  # Add to the CVC list
            print(f"Company with CVC: {company}")
            print(title)
            print(link)
            print()
            found = True

    # If no CVC-related information was found, add to the no-results list
    if not found:
        companies_with_no_results.append(company)

    print('-----------------------------')  # Divider after each company

    time.sleep(2)  # Sleep to prevent being blocked by Google for rapid requests

driver.close()

# Print the list of companies with no CVC-related results
if companies_with_no_results:
    print("Companies with no CVC results found:")
    for company in companies_with_no_results:
        print(company)
```
